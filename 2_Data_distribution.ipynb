```python
data.head()
```

```python
plt.figure(figsize=(14,6))
sns.histplot(data=data,x="fare",kde=True,hue="class",bins=50)
```

```python
df_first=data[data["class"]=="First"]
sns.histplot(data=df_first,x="fare",hue="class",kde=True)
```

```python
stats.skew(df_second["fare"])
```

```python
from scipy.stats import skew, kurtosis
```

```python
# generating example data
np.random.seed(0)
data=np.random.normal(0,1,1000)   # normal distribution
# calculating the skewness and kurtosis
data_skewness=skew(data)
data_kurtosis=kurtosis(data,fisher=True)
print(f"skewness of the dataset: {data_skewness}")
print(f"kurtosis of the dataset: {data_kurtosis}")
```

```python
sns.histplot(data,kde=True)
```

```python
# how to test the normal distribution of the data

# H0: data is normally distributed
# H1: data is not normally distributed

# shpiro wilk
# D'Agostino's K^2 Test
# Anderson-Darling Test
```

```python
# shapiro wilk test to test the normality of the data
stats.shapiro(data)
print(f"shapiro wilk test statistics: {stats.shapiro(data)[0]}")
print(f"shapiro wilk test p-value: {stats.shapiro(data)[1]}")
```
